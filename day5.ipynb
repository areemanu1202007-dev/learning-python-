```python
marks = [45, 78, 92, 60, 55]

marks.append(88)        # add at end
marks.insert(0, 99)     # add at position 0
marks.remove(60)        # remove this value
marks.pop()             # remove last item

print(marks)
print(sorted(marks))    # sorted copy
```

```python
students = ["Rahul", "Priya", "Amit", "Neha"]

for student in students:
    print("Hello " + student)

# loop with index
for i in range(len(students)):
    print(i, students[i])
```

```python
numbers = []
for i in range(5):
    num = int(input("Enter number " + str(i+1) + ": "))
    numbers.append(num)

print("Total:", sum(numbers))
print("Average:", sum(numbers)/len(numbers))
print("Highest:", max(numbers))
print("Lowest:", min(numbers))
```
