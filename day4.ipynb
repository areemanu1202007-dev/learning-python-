```python
num1 = int(input("Enter number 1: "))
num2 = int(input("Enter number 2: "))
num3 = int(input("Enter number 3: "))

sum1 = num1 + num2 + num3
sum2 = 0

# Case 1: All three numbers are identical (e.g., 4, 4, 4) -> All ignored
if num1 == num2 == num3:
    sum2 = 0

# Case 2: num1 and num2 are duplicates (e.g., 3, 3, 5) -> Only take num3
elif num1 == num2:
    sum2 = num3

# Case 3: num1 and num3 are duplicates (e.g., 3, 5, 3) -> Only take num2
elif num1 == num3:
    sum2 = num2

# Case 4: num2 and num3 are duplicates (e.g., 5, 3, 3) -> Only take num1
elif num2 == num3:
    sum2 = num1

# Case 5: All numbers are unique (e.g., 2, 3, 4) -> Take the total sum
else:
    sum2 = sum1

print("Sum of all numbers:", sum1)
print("Sum of non-duplicate numbers:", sum2)
```
